### Probability theory

Сгенерировать выборку из распределения 

$F(x)=
\begin{cases}
  0, x<0;
  \\
  x^2, x\in[0,1];
  \\
  1, x>1
\end{cases}$

с помощью равномерного распределения на отрезке $[0,1]$.

```python
import numpy as np
from numpy import random as r
import matplotlib.pyplot as plt
```

```python
def G(x):
    if x < 0:
        return 0
    elif x > 1:
        return 1
    else:
        return x
```

```python
def g(x):
    if x < 1 and x > 0:
        return 1
    else:
        return 0
```

```python
def F(x):
    if x < 0:
        return 0
    elif x > 1:
        return 1
    else:
        return x**2
```

```python
def f(x):
    if x < 1 and x > 0:
        return 2*x
    else:
        return 0
```

```python
x = np.linspace(-2, 2, 1000)

plt.figure(figsize=(16,6))

plt.subplot(121)
plt.xlabel('x')
plt.title('Given')
plt.grid(True)
plt.plot(x, np.array(list(map(G, x))), 'c-')
plt.plot(x, np.array(list(map(g, x))), 'r-')

plt.subplot(122)
plt.xlabel('x')
plt.title('Required')
plt.grid(True)
plt.plot(x, np.array(list(map(F, x))), 'c-')
plt.plot(x, np.array(list(map(f, x))), 'r-')

plt.show()
```

$F^{-1}(y)=\sqrt{y}$, где $y\sim\mathcal{U}[0,1]$, имеет требуемое распределение F

```python
y = r.uniform(0, 1, 100000)
x_req = np.sqrt(y)
```

```python
plt.figure(figsize=(16,6))

plt.subplot(121)
plt.xlabel('x')
plt.title('Required')
plt.grid(True)
plt.plot(x, np.array(list(map(f, x))), 'r-')
plt.hist(x_req, bins=30, density=True, stacked=False, color="skyblue")

plt.subplot(122)
plt.xlabel('x')
plt.title('Required')
plt.grid(True)
plt.plot(x, np.array(list(map(F, x))), 'r-')
plt.hist(x_req, bins=30, density=True, cumulative=True, color="skyblue")

plt.show()
```
